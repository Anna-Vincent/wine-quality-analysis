```python
df[df['alcohol'] > 11] #Wines with alcohol greater than 11
```

```python
df[(df['pH'] > 3.2) & (df['pH'] < 3.6)] #Wines with pH between 3.2 and 3.6
```

# **Graphical Representations**

**Distribution** **Plot**

```python
sns.histplot(df['alcohol'], kde=True) #Distribution Plot (for a single variable)
plt.title('Alcohol Distribution')
plt.show()
```

```python
sns.histplot(df['citric acid'], kde=False)
plt.title('Citric acid Distribution')
plt.show()
```

**Bar Chart**

```python
df['quality'].value_counts() #Frequency of Target Variable (quality) with bar chart

sns.countplot(x='quality', data=df)
plt.title("Wine Quality Counts")
plt.show()
```

**Box** **Plot**

```python
sns.boxplot(x='quality', y='alcohol', data=df) #Box Plot (for outliers and distribution)
plt.title('Alcohol by Wine Quality')
plt.show()
```

**Line** **Chart**

```python
df.groupby('quality')['alcohol'].mean().plot(kind='line', marker='o') #Line Plot (for trends across values)
plt.title('Average Alcohol by Quality')
plt.ylabel('Alcohol')
plt.xlabel('Quality')
plt.grid()
plt.show()
```

**Heatmap**

```python
plt.figure(figsize=(10, 6))  #Heatmap (correlation matrix)
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

**Pie** **Chart**

```python
quality_counts = df['quality'].value_counts().sort_index() # Count quality values

plt.figure(figsize=(6, 6)) # Plot pie chart
plt.pie(quality_counts, labels=quality_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Wine Quality Distribution')
plt.show()
```

**Scatter** **Plot**

```python
sns.scatterplot(x='alcohol', y='quality', data=df)  #Scatter Plot (for 2-variable relationships)
plt.title('Alcohol vs Quality')
plt.show()
```

```python
sns.scatterplot(x='pH', y='quality', data=df)  #Scatter Plot (for 2-variable relationships)
plt.title('pH vs Quality')
plt.show()
```

**Histogram**

```python
df.hist(bins=15, figsize=(15, 10), layout=(4, 3))
plt.tight_layout()
plt.show()
```
